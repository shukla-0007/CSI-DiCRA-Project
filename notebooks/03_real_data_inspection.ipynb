```python
import os
import pandas as pd

# Define the paths to your new folders
raw_path = "../data/raw"
folders = ['LST_2018', 'NDVI_2018_1', 'NDVI_2018_2', 'SSM_2018']

for folder in folders:
    folder_path = os.path.join(raw_path, folder)
    files = os.listdir(folder_path)
    
    # Filter out hidden files like .DS_Store
    valid_files = [f for f in files if not f.startswith('.')]
    
    print(f"--- Checking Folder: {folder} ---")
    print(f"Total files found: {len(valid_files)}")
    if len(valid_files) > 0:
        print(f"Example file: {valid_files[0]}")
        
        # If it's a CSV, let's look at the columns
        if valid_files[0].endswith('.csv'):
            sample_df = pd.read_csv(os.path.join(folder_path, valid_files[0]))
            print("Columns found:", sample_df.columns.tolist())
    print("\n") 
```
