```python
import os
import geopandas as gpd
import pandas as pd
import glob

# Paths to your VECTOR folders
raw_path = "../data/raw"
sources = {
    'LST': 'LST_2018/VECTOR/DISTRICT',
    'SSM': 'SSM_2018/VECTOR/DISTRICT',
    'NDVI': 'NDVI_2018_1/VECTOR/DISTRICT'
}

def extract_from_vector(label, folder_path):
    all_data = []
    files = glob.glob(os.path.join(raw_path, folder_path, "*.geojson"))
    
    print(f"Reading {len(files)} vector files for {label}...")
    
    for f in files:
        # Date is the filename (e.g., 31-01-2018.geojson)
        date_str = os.path.basename(f).replace('.geojson', '')
        
        # Load the GeoJSON
        gdf = gpd.read_file(f)
        
        # We need to find which column has the value and which has the district name
        # Based on DiCRA's format, the value is usually in 'mean' or 'value'
        # Let's find the numeric column
        val_col = gdf.select_dtypes(include=['number']).columns[0]
        name_col = 'distname' if 'distname' in gdf.columns else gdf.columns[0]
        
        for idx, row in gdf.iterrows():
            all_data.append({
                'Date': date_str,
                'District': row[name_col],
                label: row[val_col]
            })
            
    return pd.DataFrame(all_data)

# Run for all
lst_df = extract_from_vector('LST', sources['LST'])
ssm_df = extract_from_vector('SSM', sources['SSM'])
ndvi_df = extract_from_vector('NDVI', sources['NDVI'])

# Merge
master_df = pd.merge(lst_df, ssm_df, on=['Date', 'District'], how='outer')
master_df = pd.merge(master_df, ndvi_df, on=['Date', 'District'], how='outer')

# Format Date
master_df['Date'] = pd.to_datetime(master_df['Date'], dayfirst=True)
master_df = master_df.sort_values(['District', 'Date'])

print("\n✅ SUCCESS: Vector-based Master Table created!")
display(master_df.head(10))

master_df.to_csv("../data/processed/real_2018_master.csv", index=False) 
```
