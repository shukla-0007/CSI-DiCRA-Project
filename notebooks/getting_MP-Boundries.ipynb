```python
import geopandas as gpd
import os
import zipfile

# 1. Load your existing GeoJSON
gdf = gpd.read_file("../data/mp_boundaries.json")

# 2. Save it as a Shapefile locally (it creates multiple files)
temp_dir = "../data/temp_shp"
os.makedirs(temp_dir, exist_ok=True)
gdf.to_file(os.path.join(temp_dir, "mp_districts.shp"))

# 3. Zip all the components (.shp, .shx, .dbf, .prj) into one file
with zipfile.ZipFile("../data/mp_boundaries_for_gee.zip", 'w') as zipf:
    for file in os.listdir(temp_dir):
        zipf.write(os.path.join(temp_dir, file), file)

print("✅ Success! Go to your 'data' folder and upload 'mp_boundaries_for_gee.zip' to GEE.") 
```

```python
import geopandas as gpd
import pandas as pd

# 1. Load the boundaries you already have
districts = gpd.read_file("../data/mp_boundaries.json")

# 2. Calculate the center points
# Note: NAME_2 is the column for district names if you used GADM
name_col = 'district_name' if 'district_name' in districts.columns else 'NAME_2'

districts['centroid'] = districts.geometry.centroid
df_coords = pd.DataFrame({
    'district_name': districts[name_col],
    'lat': districts.centroid.y,
    'lon': districts.centroid.x
})

# 3. Save it as a REAL CSV
df_coords.to_csv("../data/mp_district_coordinates.csv", index=False)
print("✅ Created: data/mp_district_coordinates.csv")
```

```python
import geopandas as gpd

# 1. Load the India GADM file
india = gpd.read_file("../data/gadm41_IND_2.json")

# 2. DEBUG: Let's see exactly how they spelled "Madhya Pradesh"
print("Available States in file:")
print(india['NAME_1'].unique())

# 3. Filter (The string must match EXACTLY what was printed above)
state_name = "MadhyaPradesh" 
mp = india[india['NAME_1'] == state_name].copy()

if len(mp) > 0:
    mp = mp[['NAME_2', 'geometry']]
    mp.columns = ['district_name', 'geometry']
    mp.to_file("../data/mp_boundaries.json", driver='GeoJSON')
    print(f"✅ Success! Saved {len(mp)} districts to mp_boundaries.json")
else:
    print("❌ Still 0 districts. Check the spelling in the list above!") 
```
