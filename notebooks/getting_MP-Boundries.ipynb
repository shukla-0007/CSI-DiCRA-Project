```python
import geopandas as gpd

# 1. Load the India GADM file
india = gpd.read_file("../data/gadm41_IND_2.json")

# 2. DEBUG: Let's see exactly how they spelled "Madhya Pradesh"
print("Available States in file:")
print(india['NAME_1'].unique())

# 3. Filter (The string must match EXACTLY what was printed above)
state_name = "MadhyaPradesh" 
mp = india[india['NAME_1'] == state_name].copy()

if len(mp) > 0:
    mp = mp[['NAME_2', 'geometry']]
    mp.columns = ['district_name', 'geometry']
    mp.to_file("../data/mp_boundaries.json", driver='GeoJSON')
    print(f"✅ Success! Saved {len(mp)} districts to mp_boundaries.json")
else:
    print("❌ Still 0 districts. Check the spelling in the list above!") 
```
